1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
import numpy as np

def binary(num,dec = False, hex = False):
    if(dec):
        lis = []
        num1 = num
        while(num1>0):
            cif = num1 % 2
            lis.append(cif)
            num1 = num1//2
        res = 0
        for idx,j in enumerate(lis):
            res += j*10**idx
        if hex:
            print(f"Il numero in esadecimale è: {num} , mentre il numero in binario {res}")
        else:
            print(f"Il numero in decimale è: {num} , mentre il numero in binario {res}")
            
        return(res)
    
    elif(hex):
        res=0
        for idx,val in enumerate(num[::-1],0):
            if val=="A":
                res += 10*16**idx
            if val=="B":
                res += 11*16**idx
            if val=="C":
                res += 12*16**idx
            if val=="D":
                res += 13*16**idx
            if val=="E":
                res += 14*16**idx
            if val=="F":
                res += 15*16**idx
            if(val.isnumeric()):
                res += int(val)*16**idx
        binary(res,dec=True,hex=True)
             

            
```

```python
def decimal(num,binary=False):
    res=0
    if binary:
        for idx, val in enumerate(str(num)):
            res+= int(val)*2**idx
        return(res)
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def bittofloat(num):
    var = str(num)
    mantissa = 1
    bias = 127
    for idx,val in enumerate(var[9:],1) :
        mantissa += int(val)*2**(-idx)
    exp = decimal(int(var[1:9]),binary=True)
    print(int(var[1:9]))
    res = (-1)**(int(var[0]))*mantissa*2**(exp-bias)
    
    return(res)
```
